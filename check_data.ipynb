```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open("us_census_full/columns-fixed.txt","r")
colnames = [i.split(':')[0] for i in f.readlines()]
colnames = [i.replace(' ','_').replace('-','_').replace("'",'') for i in colnames]
colnames += ['target']
train = pd.read_csv("us_census_full/census_income_learn.csv", names=colnames)
#train = pd.read_csv("us_census_full/census_income_learn.csv", names=colnames, na_values = ' ?')
del train['instance_weight']   # si j'ai bien compris
```

```python
feature_continuous = ["age", "wage_per_hour", "capital_gains", "capital_losses",
                      "dividends_from_stocks", "weeks_worked_in_year"]
feature_categorical = [i for i in train.columns if i not in feature_continuous+['target']]
```

```python
#la classe 1 ( + 50K) très minoritaire, environ 6%
print train['target'].value_counts().div(train.shape[0])
```

### Continuous features

```python
train[feature_continuous].describe()
```

```python
# age est bien une variable continue
train.hist(column='age',color='k',alpha=0.5,bins=50);
```

Les distribution des ages sont très nettement differentes pour les riches et pauvres.
Réciproquement, on s'attend à ce que l'age soit informatif sur la classification riche-pauvre.

```python
_ = train.hist(column='age',bins=40,normed=True,alpha=0.5,by='target',sharey=True)
#train.groupby('target')['age'].hist(bins=40,normed=True,alpha=0.5)
#plt.title('age');
#plt.legend()
```

```python
# une grande partie des valeurs exactement egales à zéro!
weird = feature_continuous[1:5]
print 'pourcentage de 0:' 
print (train[weird]==0).mean(axis=0).apply(lambda x: "{0:.1f}%".format(100*x))
```

Le salaire horaire des riches (vert), quand ils en touchent un, est supérieure à celui des pauvres (blue). C'est aussi vrai, mais moins marqué, pour les dividendes et gain ou perte de capitaux.

```python
# histo pour les 2-10% restant
for feature in weird:
    fig = plt.figure()
    d = train[train[feature]>0].groupby('target')[feature]
    _ = d.hist(alpha=0.5,bins=30,normed=True,figsize=(6,3))
    plt.title(feature)
```

Malheuresement, la distribution au-dessus ne concerne que très peu de point de données (2-10%).
Mais est-ce que la présence ou l'absence de dividendes, gains et perte de capitaux, et salaire horaire pourrait être informatif ?

```python
# idee: transformer ces features en une variable binaire (feature==0 ou feature>0) 
df = train[weird+['target']].copy()
for feature in weird: df[feature] = (df[feature]!=0).astype(int)
```

Pour le salaire horaire, ce n'est pas convaincant

```python
count = pd.crosstab(df.wage_per_hour, df.target).T
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
```

Par contre pour les dividendes, gains ou pertes de capitaux, c'est marquant. Les riches s'amusent plus souvent en bourse.

```python
count = pd.crosstab(df.dividends_from_stocks, df.target).T
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
count = pd.crosstab(df.capital_gains, df.target).T
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
count = pd.crosstab(df.capital_losses, df.target).T
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
```

Mais alors on peut également directement vérifier la réciproque: le fait de s'amuser en bourse est bien prédictif d'être riche ou pauvre.



```python
count = pd.crosstab(df.dividends_from_stocks, df.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
count = pd.crosstab(df.capital_gains, df.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
count = pd.crosstab(df.capital_losses, df.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
```

On pourrait d'ailleurs créer une nouvelle feature `boursicote`, qui effectivement prédit bien le fait d'être riche.

```python
df['boursicote'] =  df.dividends_from_stocks + df.capital_losses + df.capital_gains
count = pd.crosstab(df.boursicote, df.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
```

La feature `weeks_worked_in_year` est presque une variable catégorique, soit aucune semaines travaillées, soit travail toute l'année. 


```python
train.hist(column='weeks_worked_in_year',bins=50);
```

Cette feature est prédictive: très faible propotion de riches chez eux qui ne travaillent jamais, et une importante proportion chez ceux qui travaillent toute l'année.

```python
df['weeks_worked_in_year'] = pd.cut(train.weeks_worked_in_year, 
                                    [-1, 0,51,52], labels=['0','1-51','52'])
count = pd.crosstab(df.weeks_worked_in_year, df.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold'])
```

### Categorical features

```python
from matplotlib.backends.backend_pdf import PdfPages
filename = "plot_categorical_features.pdf"
pdf = PdfPages(filename)

for feature in  feature_categorical:
    ax = train[feature].value_counts().plot(kind='bar')
    fig = ax.get_figure()
    ax.set_title(feature)
    pdf.savefig(fig)
    fig.clear()
pdf.close()
```

On voit dans les `barplot` qu'il y a beaucoup de '?' ou de 'Not in universe'. 
Serait-ce des valeur NA ? Il faudrait s'en occuper dans un deuxième temps.

```python
QM = train[feature_categorical].apply(lambda x: (x==" ?").mean(), axis=0)
QM = (100*QM).round()
NIU = train[feature_categorical].apply(lambda x: (x==" Not in universe").mean(), axis=0)
NIU = (100*NIU).round()
d = pd.DataFrame( {'?' : QM, 'Not in universe' : NIU, 'categories' : 0} )
for f in feature_categorical: d.ix[f,'categories'] = train[f].unique().size
d
```

Est ce que le niveau d'éducation est prédictif du revenu ? Ah les PhD s'en sortent bien !
Aucun enfant ne gagnent pas plus de 50K.
De façon plus général, quand le niveau d'éducation augmente, la proportion de riche aussi. Le niveau d'éducation semble donc très informatif sur le revenue.

On pourrait penser à deux opérations (feature engineering) sur cette feature:
- regouper des niveux d'éducation semblables, dans le but de réduire les catégories ;
- ou ordonner par niveau d'éducation (primaire < collége < lycée < bac ...) et utiliser le rang comme une variable continue.

```python
count = pd.crosstab(train.education, train.target)
prop = count.div(count.sum(1).astype(float), axis=0)
prop.plot(kind='barh', stacked=True, color=['black','gold']);
```

### Résumé

La distribution de l'age est bien continue, et elle très différente pour les riches et pauvres.
Pour les features suivantes : dividendes, gains ou pertes de capital, nombre de semaines travaillées, salaire, la distribution est très piquée à zéro. On va transformer ces variables en features catégoriques (présence ou absence de dividendes, etc...) qui d'ailleurs semblent informatives sur le revenu.

Parmis les variables catégoriques, on a regardé le niveau d'éducation, qui semble de fait très informatif sur le revenu. 

Parmis les variables catégoriques, il y en a certaines qui ont quasiment que la valeur "Not in universe" ou "?", ce qui signifient peut-être des valeurs NA. Il faudra sans doute les éliminer de l'analyse. Il y aussi des variables avec beaucoup de catégories, par exemple `detailed_industry_recode` (52) ou `country_of_birth_father` (43). Pour alléger les calculs, je vais retirer les features avec trop de catégories.
